```python
cols = [ 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']
```

```python
cols
```

```python
for c in cols : 
    q1 = dfcard[c].quantile(0.25)
    q3 = dfcard[c].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - (1.5 * iqr)
    upper = q3 + (1.5 * iqr)
    dfcard[c] = dfcard[c].map(lambda x : lower if x < lower else  upper if x > upper else x)
```

```python
plt.figure(figsize = (12,8))
sns.boxplot(dfcard)
plt.show()
```

### imbalance

```python
y.value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler()
```

```python
ros
```

```python
X_ros, y_ros = ros.fit_resample(X, y)
```

```python
y_ros.value_counts()
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_ros, y_ros, test_size = 0.30, random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc_model = RandomForestClassifier()
```

```python
rfc_model
```

```python
rfc_model.fit(X_train, y_train)
```

```python
rfc_model.score(X_test, y_test)
```
